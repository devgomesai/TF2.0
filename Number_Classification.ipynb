```python
# Verify TensorFlow installation
import tensorflow as tf
print('TensorFlow version:', tf.__version__)

# Proceed with the classification problem setup
# For this example, we'll use the MNIST dataset, a classic dataset for image classification
from tensorflow.keras.datasets import mnist

# Load the dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Print the shape of the data to understand its structure
print('Training data shape:', x_train.shape)
print('Test data shape:', x_test.shape)
```

```python
# Normalize the data
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255

# Flatten the images for the fully connected layer
x_train = x_train.reshape((x_train.shape[0], 28 * 28))
x_test = x_test.reshape((x_test.shape[0], 28 * 28))

# Convert labels to one-hot encoding
num_classes = 10
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

# Build the model
model = keras.Sequential([
    layers.Dense(512, activation='relu', input_shape=(28 * 28,)),
    layers.Dropout(0.2),
    layers.Dense(num_classes, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

print('Model built and compiled successfully.')
```
