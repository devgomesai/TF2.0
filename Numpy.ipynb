# **Speed Test**

````python
## speed comparison
from datetime import datetime

a = np.random.randn(100)
print("a:",a)
b = np.random.randn(100)
print("b:",b)
T = 100000

def slow_dot_product(a,b):
  result= 0
  for e, f in zip(a,b):
    result += e * f
  return result

t0 = datetime.now()
for t in range(T):
  slow_dot_product(a,b)
dt1 = datetime.now() - t0
print("dt1 for loop:", dt1.total_seconds())

t0 = datetime.now()
for t in range(T):
  a.dot(b)
dt2 = datetime.now() - t0
print("dt2 for loop:", dt2.total_seconds())

print("dt1 / dt2:", dt1.total_seconds() / dt2.total_seconds())
````
