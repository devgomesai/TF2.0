# **Generate Fake Data Using Faker Library**

````python
import pandas as pd
import random
from faker import Faker

# Initialize Faker
fake = Faker()

# Number of records
num_records = 10000

# Generate random data
data = {
    "Hospital Name": [fake.company() + " Hospital" for _ in range(num_records)],
    "Patient Name": [fake.name() for _ in range(num_records)],
    "Region": [fake.state() for _ in range(num_records)],
    "Country": [fake.country() for _ in range(num_records)],
    "Patient Age": [random.randint(1, 100) for _ in range(num_records)],
    "Disease/Diagnosis": [random.choice(['Flu', 'Cold', 'Covid-19', 'Heart Disease', 'Diabetes', 'Fracture']) for _ in range(num_records)],
    "Admission Date": [fake.date_this_year() for _ in range(num_records)],
    "Discharge Date": [fake.date_this_year() for _ in range(num_records)],
    "Doctor Assigned": [fake.name() for _ in range(num_records)]
}

# Create a DataFrame
df = pd.DataFrame(data)
````
